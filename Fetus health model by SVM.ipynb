```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('fetal_health.csv')
```

```python
df=df.astype({'fetal_health':'int64'})
```

```python
sns.countplot(x=df['fetal_health'])
plt.xlabel('Fetal_Health')
plt.ylabel('Count')
plt.title("COUNT VALUE")
plt.show()
```

```python
cr=df.corr()
cr.style.background_gradient()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,0:21]
y=df['fetal_health']
```

```python
xtr,xts,ytr,yts=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
xtr
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
sca=StandardScaler()
```

```python
xtr=sca.fit_transform(xtr)
xts=sca.transform(xts)
```

```python
mod=SVC()
mod.fit(xtr,ytr)
```

```python
pred=mod.predict(xts)
print('Result:',pred)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
a=accuracy_score(yts,pred)
print("Accuracy:",a*100)
print("--------------------------------------")
b=print("Classification Report is: \n",classification_report(yts,pred))
print('--------------------------------------')
c=confusion_matrix(yts,pred)
print("Confusion Matrix is: \n",c)
```

```python
sns.heatmap(c,annot=True,fmt='g',cmap='summer',
            xticklabels=['1','2','3'],yticklabels=['1','2','3'])
plt.xlabel('Actual',fontsize=16)
plt.ylabel('Predicted',fontsize=16)
plt.title("Confusion Matrix",fontsize=20)
```
